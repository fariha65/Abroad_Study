```python
def prog_details(soup):
    # Find the "Quick Facts" section
    quick_facts = soup.find("div", {"id": "mini-panel-programs_quick_facts"})

    # Check if quick_facts was found, return empty dictionary if not found
    if quick_facts is None:
        print("Quick facts section not found.")
        return {}

    quick_facts_panels = quick_facts.find_all("div", "panel-pane")
    facts = [None] * len(quick_facts_panels)

    for i in range(len(quick_facts_panels)):
        if i == 2:
            facts[i] = quick_facts_panels[i].find('div', 'field')
        else:
            facts[i] = quick_facts_panels[i].find('span', 'quick-facts-linked-content')

    # Create a dictionary with collected data, handling missing elements
    program_details = {
        "Faculty": facts[0].text.strip() if len(facts) > 0 and facts[0] else "Not available",
        "Duration": facts[1].text.strip() if len(facts) > 1 and facts[1] else "Not available",
        "Program_Components": facts[2].text.strip() if len(facts) > 2 and facts[2] else "Not available",
        "Academic_Unit": facts[3].text.strip() if len(facts) > 3 and facts[3] else "Not available",
        "Specialization": facts[4].text.strip() if len(facts) > 4 and facts[4] else "Not available",
        "Tuition_Fees": facts[5].text.strip() if len(facts) > 5 and facts[5] else "Not available",
    }

    return program_details
```

```python
import requests
from bs4 import BeautifulSoup

#req = requests.get("https://www.grad.ubc.ca/prospective-students/graduate-degree-programs/master-science-craniofacial-science-diploma-endodontics")
req = requests.get('https://www.grad.ubc.ca/prospective-students/graduate-degree-programs/master-of-science-zoology')
soup = BeautifulSoup(req.content, "html.parser")

#  submission information
panel_div = soup.find('div', class_='panel-pane pane-custom pane-8')

if panel_div:
    # Application submission info
    try:
        english_test_submission_time = panel_div.find('p').find('strong').get_text(strip=True)
        print(f"English language test application Submission time: {english_test_submission_time}")
    except AttributeError:
        print("English language test application Submission time is not found.")

    # Handle TOEFL section
    try:
        toefl_header = panel_div.find('h5', string="TOEFL: Test of English as a Foreign Language - internet-based")
        if toefl_header:
            # Find TOEFL overall score
            try:
                toefl_overall_score_requirement = toefl_header.find_next('p').get_text(strip=True).split(":")[1].strip()
                print(f"TOEFL Overall Score Requirement: {toefl_overall_score_requirement}")
            except (AttributeError, IndexError):
                print("TOEFL Overall Score Requirement is not found.")

            # Find TOEFL section-wise scores
            try:
                toefl_scores_div = toefl_header.find_next('div', class_='columns-4 clearfix row-fluid gpsfourcolumns')
                if toefl_scores_div:
                    span3_divs = toefl_scores_div.find_all('div', class_='span3')
                    for div in span3_divs:
                        toefl_score_requirement_label = div.find('strong').get_text(strip=True)
                        toefl_score_requirement = div.find_all('p')[1].get_text(strip=True)
                        print(f"{toefl_score_requirement_label} TOEFL Score Requirement: {toefl_score_requirement}")
                else:
                    print("TOEFL section scores  is not available.")
            except (AttributeError, IndexError):
                print("Error finding TOEFL section scores.")
        else:
            print("TOEFL header not found.")
    except Exception as e:
        print(f"Error handling TOEFL section: {e}")

    #  IELTS
    try:
        ielts_header = panel_div.find('h5', string="IELTS: International English Language Testing System")
        if ielts_header:
            # Find IELTS overall score
            try:
                ielts_overall_score_requirement = ielts_header.find_next('p').get_text(strip=True).split(":")[1].strip()
                print(f"IELTS Overall Score Requirement: {ielts_overall_score_requirement}")
            except (AttributeError, IndexError):
                print("IELTS Overall Score Requirement is not found.")

            # Find IELTS section-wise scores
            try:
                ielts_scores_div = ielts_header.find_next('div', class_='columns-4 clearfix row-fluid gpsfourcolumns')
                if ielts_scores_div:
                    span3_divs = ielts_scores_div.find_all('div', class_='span3')
                    for div in span3_divs:
                        ielts_score_requirement_label = div.find('strong').get_text(strip=True)
                        ielts_score_requirement = div.find_all('p')[1].get_text(strip=True)
                        print(f"{ielts_score_requirement_label} IELTS Score Requirement: {ielts_score_requirement}")
                else:
                    print("IELTS section scores div not found.")
            except (AttributeError, IndexError):
                print("Error finding IELTS section scores.")
        else:
            print("IELTS header not found.")
    except Exception as e:
        print(f"Error handling IELTS section: {e}")

    # GRE
    try:
        gre_panel_div = soup.find('div', class_='panel-pane pane-custom pane-9')
        if gre_panel_div:
            try:
                p_tags = gre_panel_div.find_all('p')
                if len(p_tags) > 1:
                    gre_requirement = p_tags[1].get_text(strip=True)
                    print(f"GRE Score Requirement: {gre_requirement}")
                else:
                    print("Not enough <p> tags found in GRE section.")
            except (AttributeError, IndexError):
                print("Error finding GRE requirement details.")
        else:
            print("GRE requirement not available.")
    except Exception as e:
        print(f"Error handling GRE section: {e}")
else:
    print("Panel div for the application submission and test scores not found.")
```

```python

def get_test_scores(soup):
    scores = {}

    panel_div = soup.find('div', class_='panel-pane pane-custom pane-8')

    if panel_div:
        # Application submission info
        try:
            english_test_submission_time = panel_div.find('p').find('strong').get_text(strip=True)
            scores['english_test_submission_time'] = english_test_submission_time
        except AttributeError:
            scores['english_test_submission_time'] = "Not found"

        # Handle TOEFL section
        try:
            toefl_header = panel_div.find('h5', string="TOEFL: Test of English as a Foreign Language - internet-based")
            if toefl_header:
                toefl_overall_score_requirement = toefl_header.find_next('p').get_text(strip=True).split(":")[1].strip()
                scores['toefl_overall_score_requirement'] = toefl_overall_score_requirement

                toefl_scores_div = toefl_header.find_next('div', class_='columns-4 clearfix row-fluid gpsfourcolumns')
                if toefl_scores_div:
                    span3_divs = toefl_scores_div.find_all('div', class_='span3')
                    toefl_scores = {}
                    for div in span3_divs:
                        toefl_score_requirement_label = div.find('strong').get_text(strip=True)
                        toefl_score_requirement = div.find_all('p')[1].get_text(strip=True)
                        toefl_scores[toefl_score_requirement_label] = toefl_score_requirement
                    scores['toefl_scores'] = toefl_scores
                else:
                    scores['toefl_scores'] = "Not found"
            else:
                scores['toefl_overall_score_requirement'] = "Not found"
        except Exception as e:
            scores['toefl_scores'] = "Error handling TOEFL section"

        # Handle IELTS section similarly
        try:
            ielts_header = panel_div.find('h5', string="IELTS: International English Language Testing System")
            if ielts_header:
                ielts_overall_score_requirement = ielts_header.find_next('p').get_text(strip=True).split(":")[1].strip()
                scores['ielts_overall_score_requirement'] = ielts_overall_score_requirement

                ielts_scores_div = ielts_header.find_next('div', class_='columns-4 clearfix row-fluid gpsfourcolumns')
                if ielts_scores_div:
                    span3_divs = ielts_scores_div.find_all('div', class_='span3')
                    ielts_scores = {}
                    for div in span3_divs:
                        ielts_score_requirement_label = div.find('strong').get_text(strip=True)
                        ielts_score_requirement = div.find_all('p')[1].get_text(strip=True)
                        ielts_scores[ielts_score_requirement_label] = ielts_score_requirement
                    scores['ielts_scores'] = ielts_scores
                else:
                    scores['ielts_scores'] = "Not found"
            else:
                scores['ielts_overall_score_requirement'] = "Not found"
        except Exception as e:
            scores['ielts_scores'] = "Error handling IELTS section"

        # Handle GRE similarly
        try:
            gre_panel_div = soup.find('div', class_='panel-pane pane-custom pane-9')
            if gre_panel_div:
                p_tags = gre_panel_div.find_all('p')
                if len(p_tags) > 1:
                    scores['gre_requirement'] = p_tags[1].get_text(strip=True)
                else:
                    scores['gre_requirement'] = "Not available"
            else:
                scores['gre_requirement'] = "Not available"
        except Exception as e:
            scores['gre_requirement'] = "Error handling GRE section"
    else:
        scores = "Panel div for application submission and test scores not found."

    return scores
```

```python
def get_intakes(soup):
    intake_data = {}

    admission_tab = soup.find('div', id='ga-Admission')
    if admission_tab:
        meet_deadline = admission_tab.find_all('div')[1]

        # Handling September 2025 Intake
        views_row_1_divs = meet_deadline.find('div', class_='views-row-first')
        if views_row_1_divs:
            intake_data["September_2025"] = {
                "application_open": {},
                "canadian": {},
                "international": {}
            }

            application_open_div = views_row_1_divs.find('div', 'views-field-ipo-beg2')
            if application_open_div:
                application_label = application_open_div.find('h5').text.strip()
                application_date = application_open_div.find('span').text.strip()
                intake_data["September_2025"]["application_open"] = {
                    "label": application_label,
                    "date": application_date
                }

            # Canadian Applicants deadlines
            canadian_deadlines = ["ipo-udf1", "ipo-udf5", "ipo-udf7"]
            for deadline in canadian_deadlines:
                div = views_row_1_divs.find('div', f'views-field-{deadline}')
                if div:
                    label = div.find('strong').text.strip()
                    date = div.find('span').text.strip()
                    intake_data["September_2025"]["canadian"][label] = date

            # International Applicants deadlines
            international_deadlines = ["ipo-udf1", "ipo-udf5", "ipo-udf7"]
            for deadline in international_deadlines:
                div = views_row_1_divs.find('div', f'views-field-{deadline}')
                if div:
                    label = div.find('strong').text.strip()
                    date = div.find('span').text.strip()
                    intake_data["September_2025"]["international"][label] = date

        # Handling January 2026 Intake
        views_row_2_divs = meet_deadline.find('div', class_='views-row-2')
        if views_row_2_divs:
            intake_data["January_2026"] = {
                "application_open": {},
                "canadian": {},
                "international": {}
            }

            application_open_div = views_row_2_divs.find('div', 'views-field-ipo-beg2')
            if application_open_div:
                application_label = application_open_div.find('h5').text.strip()
                application_date = application_open_div.find('span').text.strip()
                intake_data["January_2026"]["application_open"] = {
                    "label": application_label,
                    "date": application_date
                }

            # Canadian Applicants deadlines
            for deadline in canadian_deadlines:
                div = views_row_2_divs.find('div', f'views-field-{deadline}')
                if div:
                    label = div.find('strong').text.strip()
                    date = div.find('span').text.strip()
                    intake_data["January_2026"]["canadian"][label] = date

            # International Applicants deadlines
            for deadline in international_deadlines:
                div = views_row_2_divs.find('div', f'views-field-{deadline}')
                if div:
                    label = div.find('strong').text.strip()
                    date = div.find('span').text.strip()
                    intake_data["January_2026"]["international"][label] = date

    return intake_data
```

```python
def get_application(soup):
    application_div = soup.find('div', id='mini-panel-programs_admission_prepare_appli')
    application_info = []

    if application_div:
        application_divs = application_div.find_all('div', class_='panel-pane')
        if application_divs:
            for div in application_divs:
                Prepare_Application_Info = div.find('div', class_='field-label-hidden')
                if Prepare_Application_Info:
                    application_info.append(Prepare_Application_Info.text.strip())
                else:
                    application_info.append(div.text.strip())
        else:
            return ["No panel-pane divs found."]
    else:
        return ["No element found."]

    return application_info  # Return the collected information as a list
```

```python
def get_tuition(soup):
    tuition_info = {}  # Dictionary to store all the tuition-related information

    try:
        # Locate the Tuition table
        Tution_table = soup.find('table')
        if not Tution_table:
            raise Exception("Tuition table not found.")

        # Table headers (thead)
        thead = Tution_table.find('thead')
        if not thead:
            raise Exception("Table header (thead) not found.")

        tution_table_titles = thead.find_all('th')
        col_1_title = tution_table_titles[0].text.strip() if len(tution_table_titles) > 0 else 'Title not found'
        col_2_title = tution_table_titles[1].text.strip() if len(tution_table_titles) > 1 else 'Title not found'
        col_3_title = tution_table_titles[2].text.strip() if len(tution_table_titles) > 2 else 'Title not found'

        # Table body (tbody)
        tbody = Tution_table.find('tbody')
        if not tbody:
            raise Exception("Table body (tbody) not found.")

        # Get all rows in the table body
        tution_table_rows = tbody.find_all('tr')
        if len(tution_table_rows) == 0:
            raise Exception("No rows found in the table body.")

        # 1st row - Application fees
        application_fees = tution_table_rows[0].find_all('td')
        tuition_info['Application_Fees'] = {
            'Domestic': application_fees[1].text.strip() if len(application_fees) > 1 else 'Value not found',
            'International': application_fees[2].text.strip() if len(application_fees) > 2 else 'Value not found'
        }

        # 3rd row - Installments per year
        if len(tution_table_rows) > 2:
            tution_fees_installments = tution_table_rows[2].find_all('td')
            tuition_info['Installments_Per_Year'] = {
                'Domestic': tution_fees_installments[1].text.strip() if len(tution_fees_installments) > 1 else 'Value not found',
                'International': tution_fees_installments[2].text.strip() if len(tution_fees_installments) > 2 else 'Value not found'
            }

        # 4th row - Tuition per installment
        if len(tution_table_rows) > 3:
            tuition_per_installment = tution_table_rows[3].find_all('td')
            tuition_info['Tuition_Per_Installment'] = {
                'Domestic': tuition_per_installment[1].text.strip() if len(tuition_per_installment) > 1 else 'Value not found',
                'International': tuition_per_installment[2].text.strip() if len(tuition_per_installment) > 2 else 'Value not found'
            }

        # 5th row - Tuition per year
        if len(tution_table_rows) > 4:
            tuition_per_year = tution_table_rows[4].find_all('td')
            tuition_info['Tuition_Per_Year'] = {
                'Domestic': tuition_per_year[1].text.strip() if len(tuition_per_year) > 1 else 'Value not found',
                'International': tuition_per_year[2].text.strip() if len(tuition_per_year) > 2 else 'Value not found'
            }

        # 6th row - International Tuition Award per year
        if len(tution_table_rows) > 5:
            ita_per_year = tution_table_rows[5].find_all('td')
            tuition_info['International_Tuition_Award_Per_Year'] = {
                'Domestic': ita_per_year[1].text.strip() if len(ita_per_year) > 1 else 'Value not found',
                'International': ita_per_year[2].text.strip() if len(ita_per_year) > 2 else 'Value not found'
            }

        # 8th row - Student Fees per year
        if len(tution_table_rows) > 7:
            student_fees = tution_table_rows[7].find_all('td')
            tuition_info['Student_Fees_Per_Year'] = student_fees[1].text.strip() if len(student_fees) > 1 else 'Value not found'

    except Exception as e:
        tuition_info['error'] = str(e)

    return tuition_info  # Return the dictionary containing all tuition information
```

```python
def get_apply(soup):
    # Initialize a dictionary to store application links
    application_info = {
        "admission_information_requirements": None,
        "application_links": []
    }

    # Find the main admission information div
    admission_information_requirements = soup.find('div', id='mini-panel-programs_admission_information_a')
    if admission_information_requirements:
        # Find the last div within the admission information section
        apply_divs = admission_information_requirements.find_all('div')[-1] if admission_information_requirements.find_all('div') else None

        if apply_divs:
            # Find all application links within the div
            application_links = apply_divs.find_all('a')
            if application_links:
                # Collect application form names and URLs in the dictionary
                for link in application_links:
                    application_info["application_links"].append({
                        "Application_Form": link.get_text(),
                        "Application_link": link.get('href')
                    })
            else:
                application_info["application_links"] = "No application links found."
        else:
            application_info["admission_information_requirements"] = "Apply divs not found or empty."
    else:
        application_info["admission_information_requirements"] = "Admission information requirements div not found."

    return application_info
```

```python
import requests
from bs4 import BeautifulSoup

def get_further_information(soup):
    further_info = {}

    # Calendar URL
    calendar_div = soup.find('div', class_='panel-pane pane-entity-field pane-node-field-prog-calendar-link')
    if calendar_div:
        calendar_url = calendar_div.find('a', href=True)
        further_info['calendar_url'] = calendar_url['href'] if calendar_url else "No link found"
    else:
        further_info['calendar_url'] = "Calendar div with class not found"

    # Program Website
    prog_website_div = soup.find('div', class_='panel-pane pane-entity-field pane-node-field-prog-website')
    if prog_website_div:
        prog_website = prog_website_div.find('a', href=True)
        further_info['prog_website'] = {
            "name": prog_website.get_text() if prog_website else "No link found",
            "url": f"https://grad.ubc.ca{prog_website['href']}" if prog_website else "No link found"
        }
    else:
        further_info['prog_website'] = "Program website div with class not found"

    # Faculty Overview
    f_overview_div = soup.find('div', class_='panel-pane pane-entity-field pane-taxonomy-term-field-faculty-internal-reference')
    if f_overview_div:
        faculty_overview = f_overview_div.find('a', href=True)
        further_info['Faculty_Overview'] = {
            "name": faculty_overview.get_text() if faculty_overview else "No link found",
            "url": faculty_overview['href'] if faculty_overview else "No link found"
        }
    else:
        further_info['Faculty_Overview'] = "Faculty overview div with class not found"

    # Classification
    classification_div = soup.find('div', class_='panel-pane pane-entity-field pane-node-field-research-crdc')
    if classification_div:
        classification = classification_div.find('a', href=True)
        further_info['classification'] = {
            "name": classification.get_text() if classification else "No link found",
            "url": f"https://grad.ubc.ca{classification['href']}" if classification else "No link found"
        }
    else:
        further_info['classification'] = "Classification div with class not found"

    return further_info
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException

def universitySchema(soup):
    # Set up Chrome options for headless execution
    chrome_options = Options()
    chrome_options.add_argument("--headless")  # Run Chrome in headless mode
    chrome_options.add_argument("--no-sandbox")  # Recommended for Colab
    chrome_options.add_argument("--disable-dev-shm-usage")  # Overcome limited resource problems

    # Initialize WebDriver with options
    driver = webdriver.Chrome(options=chrome_options)

    # Static data for demonstration
    name = "University of British Columbia"
    location = {
        "country": "Canada",
        "state": "British Columbia",
        "city": "Vancouver"
    }
    website = "https://www.grad.ubc.ca/"

    def get_rankings(driver, url):
        driver.get(url)
        try:
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.ID, "ranking-data-load"))
            )
            ranking_div = driver.find_element(By.ID, "ranking-data-load")
            ranking = ranking_div.find_element(By.CLASS_NAME, "rank-no").text
            score = ranking_div.find_element(By.CLASS_NAME, "rank-score").text
            return {"ranking": ranking, "score": score}
        except TimeoutException:
            print("Failed to load ranking data in time.")
            return {"ranking": None, "score": None}

    rank_data = get_rankings(driver, "https://www.topuniversities.com/world-university-rankings?search=university%20of%20british%20columbia")
    driver.quit()

    university_schema = {
        "name": name,
        "location": location,
        "website": website,
        "ranking": rank_data.get("ranking"),
        "score": rank_data.get("score")
    }

    return university_schema
```
