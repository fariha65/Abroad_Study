```python
import requests
from bs4 import BeautifulSoup
def faculty_info(base_URL):

      # Define the URL
      #url = "https://umanitoba.ca/explore/programs-of-study/computer-science-msc"
      response = requests.get(base_URL)
      soup = BeautifulSoup(response.text, 'html.parser')

      # Get teaser content sections
      teaser_content_list = soup.find_all('div', class_='teaser__content')

      # Faculty Information
      if teaser_content_list:
          teaser_content = teaser_content_list[0]
          faculty_tag = teaser_content.find('h3', class_='teaser__title')

          if faculty_tag:
              faculty = faculty_tag.get_text(strip=True)
          else:
              faculty = "Faculty information not found."

          faculty_overview_tag = teaser_content.find('p')
          if faculty_overview_tag:
              faculty_overview_lines = faculty_overview_tag.get_text(separator='\n', strip=True).split('\n')

              if len(faculty_overview_lines) >= 2:
                  output = f"Faculty: {faculty_overview_lines[0].strip()} \nFaculty overview: {faculty_overview_lines[1].strip()}"
              else:
                  output = "Incomplete information in the overview."
          else:
              output = "Faculty overview not found."
      else:
          output = "Content section not found."

      print(output)

      # Degree Information
      if len(teaser_content_list) > 1:
          teaser_content1 = teaser_content_list[1]
          degree_tag = teaser_content1.find('h3', class_='teaser__title')

          if degree_tag:
              degree = degree_tag.get_text(strip=True)
          else:
              degree = "Degree information not found."

          degree_overview_tag = teaser_content1.find('p')
          if degree_overview_tag:
              degree_overview_lines = degree_overview_tag.get_text(separator='\n', strip=True).split('\n')

              if degree_overview_lines:
                  print(f"Degree: {degree_overview_lines[0].strip()}")
              else:
                  print("Incomplete information in the overview.")
          else:
              print("Degree overview not found.")
      else:
          print("Degree content section not found.")

      # Expected Duration Information
      if len(teaser_content_list) > 2:
          teaser_content2 = teaser_content_list[2]
          expected_duration_tag = teaser_content2.find('h3', class_='teaser__title')

          if expected_duration_tag:
              expected_duration = expected_duration_tag.get_text(strip=True)
          else:
              expected_duration = "Expected duration information not found."

          expected_duration_overview_tag = teaser_content2.find('p')
          if expected_duration_overview_tag:
              expected_duration_overview_lines = expected_duration_overview_tag.get_text(separator='\n', strip=True).split('\n')

              if expected_duration_overview_lines:
                  print(f"Expected duration: {expected_duration_overview_lines[0].strip()}")
              else:
                  print("Incomplete information in the overview.")
          else:
              print("Expected duration overview not found.")
      else:
          print("Expected duration content section not found.")
```

```python
import requests
from bs4 import BeautifulSoup

def meet_deadlines(base_URL):
    response = requests.get(base_URL)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find content section by ID
    content_div = soup.find('div', id='edit-sci-comp-sci')

    if content_div:
        # Extract opening dates information
        p_tags = content_div.find_all('p')
        if len(p_tags) > 1:
            a_tag = p_tags[1].find('a')
            if a_tag and 'href' in a_tag.attrs:
                print(f"Applications open date: {a_tag.get_text(strip=True)}")
                print(f"Application More detail: {a_tag['href']}")
            else:
                print("Link not found in the second <p> tag.")
        else:
            print("Not enough <p> tags found.")

        # Process Domestic applicants - Fall deadlines
        tables = content_div.find_all('table')
        if tables and len(tables) > 0:
            tbody = tables[0].find('tbody')
            if tbody:
                rows = tbody.find_all('tr')
                if len(rows) > 0:
                    th_text = rows[0].find('th').get_text(strip=True) if rows[0].find('th') else "No header"
                    td_text = rows[0].find('td').get_text(strip=True) if rows[0].find('td') else "No data"
                    print(f"{th_text} Domestic applicants application deadlines: {td_text}")
                if len(rows) > 1:
                    th_text = rows[1].find('th').get_text(strip=True) if rows[1].find('th') else "No header"
                    td_text = rows[1].find('td').get_text(strip=True) if rows[1].find('td') else "No data"
                    print(f"{th_text} Domestic applicants application deadlines: {td_text}")
            else:
                print("No <tbody> found in the first table.")
        else:
            print("No table found for domestic applicants.")

        # Process International applicants - Fall deadlines
        if len(tables) > 1:
            tbody = tables[1].find('tbody')
            if tbody:
                rows = tbody.find_all('tr')
                if len(rows) > 0:
                    th_text = rows[0].find('th').get_text(strip=True) if rows[0].find('th') else "No header"
                    td_text = rows[0].find('td').get_text(strip=True) if rows[0].find('td') else "No data"
                    print(f"{th_text} International applicants application deadlines: {td_text}")
                if len(rows) > 1:
                    th_text = rows[1].find('th').get_text(strip=True) if rows[1].find('th') else "No header"
                    td_text = rows[1].find('td').get_text(strip=True) if rows[1].find('td') else "No data"
                    print(f"{th_text} International applicants application deadlines: {td_text}")
            else:
                print("No <tbody> found in the second table.")
        else:
            print("No table found for international applicants.")
    else:
        print("Content section not found.")
```

```python
import requests
from bs4 import BeautifulSoup
def apply_online(base_URL):


      # URL of the page
      #url = "https://umanitoba.ca/graduate-studies/graduate-student-admissions/apply-graduate-studies"

      # Sending a GET request to the page
      response = requests.get(base_URL)

      # Parsing the page content with BeautifulSoup
      soup = BeautifulSoup(response.content, "html.parser")

      # Finding the specific div by its class that contains the <a> tag
      div = soup.find("div", class_="panelgroup__item-content item-content wysiwyg")

      if div:
          # Find the <a> tag inside the div
          a_tag = div.find("a")

          if a_tag:
              # Extract the href (link) and the text inside the <a> tag
              link = a_tag['href']
              link_text = a_tag.get_text(strip=True)

              # Print the results
              print(f"{link_text} Link: {link}")
          else:
              print("No <a> tag found in the specified div.")
      else:
          print("Specified div not found on the page.")
```

```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

def get_university_ranking(driver, url):
    # Open the specified URL
    driver.get(url)

    # Wait until the ranking data loads
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "ranking-data-load")))

    # Locate the ranking and score elements
    ranking_div = driver.find_element(By.ID, "ranking-data-load")
    ranking = ranking_div.find_element(By.CLASS_NAME, "rank-no").text
    score = ranking_div.find_element(By.CLASS_NAME, "rank-score").text

    # Print the ranking and overall score
    print("Ranking: ", ranking)
    print("Overall Score: ", score)


```
